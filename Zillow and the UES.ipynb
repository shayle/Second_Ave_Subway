```python
import pandas as pd
import geopandas as gpd
import numpy as np
import os
import sys
import pylab as pl
import re
import urllib as request
from datetime import datetime
import seaborn as sns
import matplotlib.pyplot as plt


pl.rcParams['figure.titlesize'] = 'large'
pl.rcParams['figure.figsize'] = (20, 20)
             
% pylab inline
```

```python
MedianPricePerSqFt = pd.read_csv('Neighborhood_MedianListingPricePerSqft_AllHomes.csv')
MedianRental = pd.read_csv('Neighborhood_MedianRentalPrice_AllHomes.csv')
Rental =MedianRental[MedianRental['City']=="New York"]
PriceSqFt =MedianPricePerSqFt[MedianPricePerSqFt['City']=="New York"]
UES = PriceSqFt[PriceSqFt['RegionName']=='Upper East Side']
UESrental = Rental[Rental['RegionName']=='Upper East Side']
```

```python
# sns.set_style("darkgrid")
ax = UES.T.iloc[6:,:].plot(figsize=(9,6))
plt.xlabel('Date')
plt.ylabel('$ per Square Foot')
plt.title('Change in Median Listing Price per Square Foot for UES Homes')
plt.legend(['UES'])
plt.show()
```

```python
# sns.set_style("darkgrid")
ax = UESrental.T.iloc[6:,:].plot(figsize=(9,6),legend=False)
plt.xlabel('Date')
plt.ylabel('$')
plt.title('Change in Median Rental Price for UES Homes')
plt.legend(['UES'])
plt.show()
```
